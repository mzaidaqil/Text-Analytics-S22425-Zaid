```python
import pandas as pd
import math # use for log function
from collections import Counter
```

```python
file_path = "Processed_Reviews.csv"
df = pd.read_csv(file_path)
```

```python
#Extract the tokenized text data:

tokenized_reviews = df['tokenized'].dropna().apply(eval)
```

```python
#Function to compute the Term Frequency (TF)

def compute_tf(document):
    word_count = Counter(document)
    tf = {word: count / len(document) for word, count in word_count.items()}
    return tf
```

```python
#Function to compute the Inverse Document Frequency (IDF)

def compute_idf(documents):
    N = len(documents) # Total number of documents
    idf = {}
    all_words = set(word for doc in documents for word in doc) # Unique words
    for word in all_words:
        count = sum(1 for doc in documents if word in doc)
        idf[word] = math.log(N / count)
    return idf
```

# 📌 Feature Engineering with High TF-IDF Scores

---

## 1️⃣ Why High TF-IDF Terms Matter

**TF-IDF (Term Frequency–Inverse Document Frequency)** helps identify words that are:

- Frequent in a specific document  
- Rare across the overall corpus  

High TF-IDF terms are **discriminative features** because they capture what makes a document unique.

These features are highly useful for:

- Text classification  
- Spam detection  
- Fake review detection  
- Sentiment analysis  
- Topic classification  

---

## 2️⃣ Feature Selection Using TF-IDF

### ✔ Keep Top-N Highest TF-IDF Terms

Instead of using the entire vocabulary (which may contain thousands of words), we can:

- Select only the top **N highest-scoring terms**
- Reduce dimensionality
- Improve model speed
- Reduce overfitting

### Example Strategy

- Keep top **5,000 features**
- Keep terms with **TF-IDF > 0.1**
- Apply **statistical feature selection**

This improves:

- Training efficiency  
- Generalization performance  

---

### ✔ Remove Very Low-Scoring Terms

Very low TF-IDF scores usually indicate:

- Common words across documents  
- Low informational value  
- Noise  

Removing them:

- Reduces sparsity  
- Improves signal-to-noise ratio  
- Makes the model more stable  

---

## 3️⃣ Dimensionality Reduction Techniques

After TF-IDF transformation, the matrix is often:

- High-dimensional  
- Sparse  

You can reduce dimensionality using:

### 🔹 A. SelectKBest (Chi-Square)

Select statistically significant features relative to target labels.

Used for:
- Supervised classification

---

### 🔹 B. Truncated SVD (Latent Semantic Analysis)

Projects high-dimensional TF-IDF into a lower-dimensional semantic space.

Useful when:

- Dataset is large  
- Vocabulary is huge  
- You want compressed representation  

---

## 4️⃣ Benefits for Machine Learning

Using high TF-IDF feature selection provides:

- ✅ Faster training  
- ✅ Less memory usage  
- ✅ Reduced overfitting  
- ✅ Improved model interpretability  
- ✅ Better generalization  

---

## 5️⃣ Example Pipeline (Conceptual)

1. Clean text  
2. Apply TF-IDF vectorization  
3. Select top-N features  
4. Train classifier (Logistic Regression / SVM / Naive Bayes)  
5. Evaluate performance  

---

## 6️⃣ When Not to Rely Only on TF-IDF

TF-IDF works well for:

- Traditional machine learning models  

However, it does **not capture**:

- Word order  
- Context  
- Semantic meaning  

For deeper semantic understanding, combine with:

- Word embeddings (Word2Vec, GloVe)  
- Contextual models (BERT)  

---

## 7️⃣ Best Practice Strategy

Instead of blindly keeping highest TF-IDF terms:

- Compare performance using:
  - Full vocabulary  
  - Top-N selected features  

- Perform cross-validation  
- Monitor:
  - Precision  
  - Recall  
  - F1-score  

Feature engineering is an experimental and iterative process — not a one-shot decision.
